```python
for dataset in combine:    
    dataset.loc[ dataset['Months since Last Donation'] <= 18.5, 'Months since Last Donation'] = 0
    dataset.loc[(dataset['Months since Last Donation'] > 18.5) & (dataset['Months since Last Donation'] <= 37), 'Months since Last Donation'] = 1
    dataset.loc[(dataset['Months since Last Donation'] > 37) & (dataset['Months since Last Donation'] <= 55.5), 'Months since Last Donation'] = 2
    dataset.loc[ dataset['Months since Last Donation'] > 55.5, 'Months since Last Donation'] = 3
train_df.head(20)
```

```python
for dataset in combine:    
    dataset.loc[ dataset['Number of Donations'] <= 8, 'Number of Donations'] = 0
    dataset.loc[(dataset['Number of Donations'] > 8) & (dataset['Number of Donations'] <= 15), 'Number of Donations'] = 1
    dataset.loc[(dataset['Number of Donations'] > 15) & (dataset['Number of Donations'] <= 22), 'Number of Donations'] = 2
    dataset.loc[(dataset['Number of Donations'] > 22) & (dataset['Number of Donations'] <= 29), 'Number of Donations'] = 3
    dataset.loc[(dataset['Number of Donations'] > 29) & (dataset['Number of Donations'] <= 36), 'Number of Donations'] = 4
    dataset.loc[(dataset['Number of Donations'] > 36) & (dataset['Number of Donations'] <= 43), 'Number of Donations'] = 5
    dataset.loc[ dataset['Number of Donations'] > 43, 'Number of Donations'] = 6
train_df.head(20)
```

```python
for dataset in combine:    
    dataset.loc[ dataset['Total Volume Donated (c.c.)'] <= 2700, 'Total Volume Donated (c.c.)'] = 0
    dataset.loc[(dataset['Total Volume Donated (c.c.)'] > 2700) & (dataset['Total Volume Donated (c.c.)'] <= 5150), 'Total Volume Donated (c.c.)'] = 1
    dataset.loc[(dataset['Total Volume Donated (c.c.)'] > 5150) & (dataset['Total Volume Donated (c.c.)'] <= 7600), 'Total Volume Donated (c.c.)'] = 2
    dataset.loc[(dataset['Total Volume Donated (c.c.)'] > 7600) & (dataset['Total Volume Donated (c.c.)'] <= 10050), 'Total Volume Donated (c.c.)'] = 3
    dataset.loc[ dataset['Total Volume Donated (c.c.)'] > 10050, 'Total Volume Donated (c.c.)'] = 4
train_df.head(20)
```

```python
for dataset in combine:    
    dataset.loc[ dataset['Months since First Donation'] <= 18, 'Months since First Donation'] = 0
    dataset.loc[(dataset['Months since First Donation'] > 18) & (dataset['Months since First Donation'] <= 34), 'Months since First Donation'] = 1
    dataset.loc[(dataset['Months since First Donation'] > 34) & (dataset['Months since First Donation'] <= 50), 'Months since First Donation'] = 2
    dataset.loc[(dataset['Months since First Donation'] > 50) & (dataset['Months since First Donation'] <= 66), 'Months since First Donation'] = 3
    dataset.loc[(dataset['Months since First Donation'] > 66) & (dataset['Months since First Donation'] <= 82), 'Months since First Donation'] = 4
    dataset.loc[ dataset['Months since First Donation'] > 82, 'Months since First Donation'] = 5
train_df.head(20)
```

```python
train_df = train_df.drop(['MonLastBand'], axis=1)
train_df = train_df.drop(['NumDonationsBand'], axis=1)
train_df = train_df.drop(['TotVolumeBand'], axis=1)
train_df = train_df.drop(['MonFirstBand'], axis=1)
combine = [train_df, test_df]
train_df.head(10)
```

```python
train_df.drop(train_df.columns[[0]], axis =1, inplace=True)
idNumbers = test_df[test_df.columns[0]]
test_df.drop(test_df.columns[[0]], axis =1, inplace=True)
```

```python
X_train = train_df.drop("Made Donation in March 2007", axis=1)
Y_train = train_df["Made Donation in March 2007"]
X_test  = test_df.copy()
X_train.shape, Y_train.shape, X_test.shape
```

```python
# Random Forest

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
Y_pred = random_forest.predict(X_test)
random_forest.score(X_train, Y_train)
acc_random_forest = round(random_forest.score(X_train, Y_train) * 100, 2)
acc_random_forest
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)

coeff_df = pd.DataFrame(train_df.columns.delete(4))
coeff_df.columns = ['Feature']
coeff_df["Correlation"] = pd.Series(logreg.coef_[0])

coeff_df.sort_values(by='Correlation', ascending=False)
```
